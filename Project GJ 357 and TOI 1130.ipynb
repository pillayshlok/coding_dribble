```python
import lightkurve as lk
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
search_ffi = lk.search_tesscut("TIC 413248763")
search_tpf = lk.search_targetpixelfile("TIC 413248763")
search_lcf = lk.search_lightcurve("TIC 413248763")
```

```python
ffi_data = search_ffi[1].download(cutout_size=10)
```

```python
ffi_data.plot()
```

```python
aper = ffi_data.create_threshold_mask(threshold = 30)
ffi_data.plot(aperture_mask=aper, mask_color='r');
```

```python
ffi_lc = ffi_data.to_lightcurve(aperture_mask=aper)
ffi_lc.plot(label='TIC 413248763');
```

```python
dm = lk.DesignMatrix(ffi_data.flux[:,~aper], name = 'regressors').pca(5).append_constant()
rc = lk.RegressionCorrector(ffi_lc)
corrected_lc = rc.correct(dm)
ax = ffi_lc.plot(label='Raw light curve')
corrected_lc.plot(ax=ax, label='Corrected light curve')
```

```python
corrected_lc = ffi_lc - rc.model_lc + np.percentile(rc.model_lc.flux, 5)
corrected_lc.plot(label = 'TIC 413248763', color = 'c')
```
